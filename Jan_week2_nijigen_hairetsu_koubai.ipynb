## 【問題5】Pythonの関数化

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。
    Parameters
    ----------------
    function : function
      求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x = np.arange(x_range[0], x_range[1], x_range[2]).reshape(-1,1) 
    y = function(x) # FUNCTIONに与えられた関数を実行
    array_xy = np.concatenate([x, y], axis=1) # x、yを列結合
    gradient = (y[1:]-y[:-1]) / (x[1:]-x[:-1]) # 勾配を求める

    plt.scatter(x, y) # x,yのグラフ描画
    plt.scatter(x[:-1], gradient) # X,勾配のグラフ表示
    plt.show()

    return array_xy, gradient

def function1(array_x):
    array_y = array_x**2
    return array_y

def function2(array_x):
    array_y = 2*array_x**2 + 2**array_x
    return array_y

def function3(array_x):
    array_y = np.sin(array_x**0.5)
    return array_y

# y=x**2
array_xy, gradient = compute_gradient(function1)
print("y=x**2:{}{}".format(array_xy, gradient))
# y=2x**2 + 2**x
array_xy, gradient = compute_gradient(function2)
print("y=2x**2 + 2**x:{}{}".format(array_xy, gradient))
# y=sin(x**0.5)
array_xy, gradient = compute_gradient(function3, x_range=(0, 50.1, 0.1))
print("y=sin(x**0.5):{}{}".format(array_xy, gradient))
```

## 【問題6】最小値を求める

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。
    Parameters
    ----------------
    function : function
      求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x = np.arange(x_range[0], x_range[1], x_range[2]).reshape(-1,1) 
    array_y_min, gradient_before, gradient_after = function(x) # FUNCTIONに与えられた関数を実行

    return array_y_min, gradient_before, gradient_after

def function1(array_x):
    array_y = array_x**2
    array_y_min = np.min(array_y)
    array_y_argmin = np.argmin(array_y)
    gradient_before = (array_y[array_y_argmin-1]-array_y[array_y_argmin]) / (array_x[array_y_argmin-1]-array_x[array_y_argmin])
    gradient_after = (array_y[array_y_argmin]-array_y[array_y_argmin+1]) / (array_x[array_y_argmin]-array_x[array_y_argmin+1])
    return array_y_min, gradient_before, gradient_after

def function2(array_x):
    array_y = 2*array_x**2 + 2**array_x
    array_y_min = np.min(array_y)
    array_y_argmin = np.argmin(array_y)
    gradient_before = (array_y[array_y_argmin-1]-array_y[array_y_argmin]) / (array_x[array_y_argmin-1]-array_x[array_y_argmin])
    gradient_after = (array_y[array_y_argmin]-array_y[array_y_argmin+1]) / (array_x[array_y_argmin]-array_x[array_y_argmin+1])
    return array_y_min, gradient_before, gradient_after

def function3(array_x):
    array_y = np.sin(array_x**0.5)
    array_y_min = np.min(array_y)
    array_y_argmin = np.argmin(array_y)
    gradient_before = (array_y[array_y_argmin-1]-array_y[array_y_argmin]) / (array_x[array_y_argmin-1]-array_x[array_y_argmin])
    gradient_after = (array_y[array_y_argmin]-array_y[array_y_argmin+1]) / (array_x[array_y_argmin]-array_x[array_y_argmin+1])
    return array_y_min, gradient_before, gradient_after

# y=x**2
y_min, array_y_before, array_y_after = compute_gradient(function1)
print(y_min, array_y_before, array_y_after)
# y=2x**2 + 2**x
y_min, array_y_before, array_y_after = compute_gradient(function2)
print(y_min, array_y_before, array_y_after)
# y=sin(x**0.5)
y_min, array_y_before, array_y_after = compute_gradient(function3, x_range=(0, 50.1, 0.1))
print(y_min, array_y_before, array_y_after)
```
